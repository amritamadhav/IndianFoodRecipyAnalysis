# Data Preprocessing  
# As we saw some NA and invalid values in the features, we need to correct them before using the data to get accurate results

```python
# data pre processing for region as it has nan invalid values
data[data['region'].isnull()]
```

```python
# As region has NaN, let's check the region for Uttar pradesh state and try to repace the feasible value
data[data['state']=='Uttar Pradesh']
```

```python
#Therefore, replacing NaN to north region as per the existing data
data['region'] = data['region'].replace(np.nan, 'North')
```

```python
# checking what all data has invalid value as -1 in region
data[data['region']=='-1']
```

```python
#Looks like wherever region has -1, state also has -1 values. Therefore, replacing -1 to Common as the state and region looks to be common for these dishes
data['region'] = data['region'].replace('-1', 'Common')
```

```python
#Replacing -1 to common for state wherever region has commo values
data.loc[data['region']=='Common','state']='Common'
```

```python
#Checking for other invalid values in state
data[data['state']=='-1']
```

```python
#looking at the data, north region dish looks like to be around Punjab. Hence, replacing with that
data['state'] = np.where( ( (data['region'] == 'North') & (data['state'] == '-1' ) ), 'Punjab',data['state'])
```

```python
#looking at the data, south region dish looks like to be around Tamil Nadu. Hence, replacing with that
data['state'] = np.where( ( (data['region'] == 'South') & (data['state'] == '-1' ) ), 'Tamil Nadu',data['state'])
```

```python
# checking what all data has invalid value as -1 for Flavor profile column
data[data['flavor_profile']=='-1']
```

```python
#checking the count of data for main course and various flavor so that we know what value to replace for flavor having main course
data[(data['course']=='main course') & (data['flavor_profile']=='spicy')].count()
```

```python
#checking the count of data for snacks and various flavor so that we know what value to replace for flavor having main course
data[(data['course']=='snack') & (data['flavor_profile']=='spicy')].count()
```

```python
#as the count is highest for spicy, hence replacing -1 with spicy for main course
data['flavor_profile'] = np.where( ( (data['course'] == 'main course') & (data['flavor_profile'] == '-1' ) ), 'spicy',data['flavor_profile'])
```

```python
#as the count is highest for spicy, hence replacing -1 with spicy for snacks
data['flavor_profile'] = np.where( ( (data['course'] == 'snack') & (data['flavor_profile'] == '-1' ) ), 'spicy',data['flavor_profile'])
```

```python
final_ingredients[final_ingredients.ingredients.duplicated(keep = False)]
```

```python
final_ingredients = final_ingredients.groupby("ingredients")["count"].sum().reset_index()
final_ingredients = final_ingredients.sort_values("count", ascending = False).reset_index(drop = True)
final_ingredients.head()
```

```python
final_ingredients.ingredients.nunique()
```

```python
df = data.copy()
df.head()
```

```python
# now let's do one hot encoding to ingredients
# so here in our data set i have a comma separated ingredients so i use contain() of string to match
# our final ingredients and if it is found then i put 1 other wise 0


for i in final_ingredients.ingredients:
    df[i] = 0 # here i create a column of particular ingredients and assign all value as 0
    df.loc[df.ingredients.str.contains(i), [i]] = 1
    # above i put 1 on that column which dishes has include that particular ingredients
```

```python
df
```

```python
diet_encoding = pd.get_dummies(data.diet, drop_first=True)
diet_encoding.head()
```

```python
df = pd.concat([df, diet_encoding], axis = 1)
df.head(2)
```

```python
flavor_profile_encoding = pd.get_dummies(data.flavor_profile, drop_first=True)
flavor_profile_encoding.head(2)
```

```python
df = pd.concat([df, flavor_profile_encoding], axis = 1)
df.head(2)
```

```python
course_encoding = pd.get_dummies(data.course, drop_first = True)
course_encoding.head(2)
```

```python
df = pd.concat([df, course_encoding], axis = 1)
df.head(2)
```

```python
state_encoding = pd.get_dummies(data.state, drop_first = True)
state_encoding.head(2)
```

```python
df = pd.concat([df, state_encoding], axis = 1)
df.head(2)
```

```python
region_encoding = pd.get_dummies(data.region, drop_first = True)
region_encoding.head(2)
```

```python
df = pd.concat([df, region_encoding], axis = 1)
df.head(2)
```

```python
final_data = df.drop(columns = ['ingredients', 'diet', 'prep_time', 'cook_time',
       'flavor_profile', 'course', 'state', 'region'])

final_data.head()
```

```python
final_data.set_index("name", inplace = True)
final_data.head()
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(metric = "cosine", algorithm = "brute")
model.fit(final_data)
```

```python
picked_cuisine = final_data.index[np.random.choice(final_data.shape[0])]
picked_cuisine
```

```python
distance, cuisine_index = model.kneighbors(final_data[final_data.index == picked_cuisine], n_neighbors = 11)
```

```python
distance = distance.flatten()
cuisine_index = cuisine_index.flatten()
```

```python
print("Recommendation for ", picked_cuisine, " are following")

count = 1
for i in range(cuisine_index.size):
    if final_data.index[cuisine_index[i]] == picked_cuisine:
        continue
    print(count, ". ", final_data.index[cuisine_index[i]], " with distance ", distance[i])
    count += 1
```
